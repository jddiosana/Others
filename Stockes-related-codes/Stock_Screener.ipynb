#Setting up


```python
import pandas as pd
import numpy as np
from google.colab import output
```

```python
earnings_per_share = '/content/Earnings per share.csv'
```

```python
return_on_equity = '/content/Return on Equity.csv'
```

```python
eps_unsorted = pd.read_csv(earnings_per_share)
roe_unsorted = pd.read_csv(return_on_equity)

y0 = '2021'
y1 = '2020'
y2 = '2019'
y3 = '2018'

eps = eps_unsorted.sort_values(by = 'CODE')
roe = roe_unsorted.sort_values(by = 'CODE')
```

#Screener


```python
screener = pd.DataFrame()

screener['CODE'] = eps['CODE']

placeholder = [0.0 for i in range(len(screener['CODE']))]
screener['Y1 - Y0'] = placeholder
screener['Q3 Y1 - Y0'] = placeholder
screener['Q2 Y1 - Y0'] = placeholder
screener['Q1 Y1 - Y0'] = placeholder

screener['Y2 - Y1'] = placeholder
screener['Q3 Y2 - Y1'] = placeholder
screener['Q2 Y2 - Y1'] = placeholder
screener['Q1 Y2 - Y1'] = placeholder

screener['Y3 - Y2'] = placeholder
screener['Q3 Y3 - Y2'] = placeholder
screener['Q2 Y3 - Y2'] = placeholder
screener['Q1 Y3 - Y2'] = placeholder

screener['ROE Y1'] = placeholder
screener['ROE Y2'] = placeholder
screener['ROE Y3'] = placeholder

for i in range(len(eps['CODE'])):
  screener['Q1 Y1 - Y0'][i] = ((eps['1Q ' + y0][i] - eps['1Q ' + y1][i]) / abs(eps['1Q ' + y1][i])) * 100
  screener['Q2 Y1 - Y0'][i] = ((eps['2Q ' + y0][i] - eps['2Q ' + y1][i]) / abs(eps['2Q ' + y1][i])) * 100
  screener['Q3 Y1 - Y0'][i] = ((eps['3Q ' + y0][i] - eps['3Q ' + y1][i]) / abs(eps['3Q ' + y1][i])) * 100
  screener['Y1 - Y0'][i] = ((eps[y0][i] - eps[y1][i]) / eps[y1][i]) * 100

  screener['Q1 Y2 - Y1'][i] = ((eps['1Q ' + y1][i] - eps['1Q ' + y2][i]) / abs(eps['1Q ' + y2][i])) * 100
  screener['Q2 Y2 - Y1'][i] = ((eps['2Q ' + y1][i] - eps['2Q ' + y2][i]) / abs(eps['2Q ' + y2][i])) * 100
  screener['Q3 Y2 - Y1'][i] = ((eps['3Q ' + y1][i] - eps['3Q ' + y2][i]) / abs(eps['3Q ' + y2][i])) * 100
  screener['Y2 - Y1'][i] = ((eps[y1][i] - eps[y2][i]) / eps[y2][i]) * 100

  screener['Q1 Y3 - Y2'][i] = ((eps['1Q ' + y2][i] - eps['1Q ' + y3][i]) / abs(eps['1Q ' + y3][i])) * 100
  screener['Q2 Y3 - Y2'][i] = ((eps['2Q ' + y2][i] - eps['2Q ' + y3][i]) / abs(eps['2Q ' + y3][i])) * 100
  screener['Q3 Y3 - Y2'][i] = ((eps['3Q ' + y2][i] - eps['3Q ' + y3][i]) / abs(eps['3Q ' + y3][i])) * 100
  screener['Y3 - Y2'][i] = ((eps[y2][i] - eps[y3][i]) / eps[y3][i]) * 100

  for j in range(len(roe['CODE'])):
    if eps['CODE'][i] == roe['CODE'][j]:
      screener['ROE Y1'][i] = (roe['NI ' + y1][j] / roe['SE ' + y1][j]) * 100
      screener['ROE Y2'][i] = (roe['NI ' + y2][j] / roe['SE ' + y2][j]) * 100
      screener['ROE Y3'][i] = (roe['NI ' + y3][j] / roe['SE ' + y3][j]) * 100

output.clear()
```
