Training a model

```
# Listing the features
features = ['age','sex','cp','trestbps','thal','fbs','restecg','thalach','oldpeak','slope','ca','chol']
X = data[features]
y = data['target']

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state = 50
                                                   )
```

Predictive Analytics

```
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
acc = log_reg.score(X_train, y_train)
y_pred = log_reg.predict(X_test)
```

Accuracy of the model

```
acc
```

Confusion Matrix

```
cf = confusion_matrix(y_test, y_pred)
cf
```

Predicting on new patient

```
age = int(input('Age: '))
sex = int(input("Gender (type '1' if Male, '0' if Female): "))
cp = int(input('cp: '))
trestbps = int(input('trestbps: '))
chol = int(input('chol: '))
fbs = int(input('fbs: '))
restecg = int(input('restecg: '))
thalach = int(input('thalach: '))
exang = int(input('exang: '))
oldpeak = float(input('oldpeak: '))
slope = int(input('slope: '))
ca = int(input('ca: '))
thal = int(input('thal: '))

patient = [[age,sex,cp,trestbps,chol,fbs,restecg,thalach,exang,oldpeak,slope,ca,thal]]
if log_reg.predict(patient)  == 1:
  print('Heart Disease Detected')
if log_reg.predict(patient) == 0:
  print('No Heart Disease Detected')
```
