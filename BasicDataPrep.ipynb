# Selecting Subset of Columns

```
columns = ['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage', 'Clicked on Ad']
subset_data = data[columns]
```

```
subset_data.head()
```

# Descriptive Statistics

```
subset_data.describe()
```

```
subset_data.corr()
```

# Correlation Coefficients

```
correlation_coefficients = subset_data.corr()
correlation_coefficients
```

# Scatter Plot

```
subset_data.plot.scatter(x='Age', y='Daily Time Spent on Site')
```

# Histogram

```
subset_data['Age'].plot.hist()
```

# Correlation Heatmap

```
sns.heatmap(correlation_coefficients, annot=True, cmap='Greens')
```

# Distribution and Scatter Plots

```
sns.pairplot(subset_data)
```

# Joint Plots

```
sns.jointplot(x='Age', y='Area Income', data=subset_data, height=8)
```

# Box Plots

```
sns.catplot(x='Clicked on Ad',
            y='Daily Time Spent on Site',
            data=data,
            kind='box',
            height=10)
```
